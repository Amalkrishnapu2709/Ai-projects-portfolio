```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df=pd.read_csv("top50aAi_InfluenceTracker - Sheet1.csv")
```

```python
df.head()
```

```python
df['Platform'].value_counts()
```

```python
platform_counts = df["Platform"].value_counts()
print("\n✅ Influencers per platform:")
print(platform_counts)

plt.figure(figsize=(7, 5))
plt.bar(platform_counts.index, platform_counts.values)
plt.title("Number of AI Influencers per Platform")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title("Distribution of AI Influencers by Platform",pad=30)
plt.pie(
    platform_counts,
    labels=platform_counts.index,
    autopct="%1.1f%%",
    startangle=140
)


plt.axis("equal")  # keep circle shape
plt.show()
```

```python
engagement_counts = df["Engagement"].value_counts()

print("\n✅ Engagement Level Counts:")
print(engagement_counts)

plt.figure(figsize=(7, 5))
plt.bar(engagement_counts.index, engagement_counts.values)
plt.title("Engagement Levels Across Influencers")
plt.xlabel("Engagement Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
all_niches = []

for niche in df["Niche"].dropna():
    splitted = [n.strip() for n in str(niche).split(",")]
    all_niches.extend(splitted)

niche_counts = Counter(all_niches)

print("\n✅ Niche Popularity:")
print(niche_counts)

plt.figure(figsize=(12, 6))
plt.bar(niche_counts.keys(), niche_counts.values())
plt.title("AI Niches Mentioned Across Influencers")
plt.xlabel("Niche")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
